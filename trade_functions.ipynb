```python
import random

def encounter_function(matrix):
    # Pick a random row
    row = random.choice(range(len(matrix)))

    # Pick a random column different from the row
    possible_cols = [i for i in range(len(matrix[0])) if i != row]
    col = random.choice(possible_cols)

    return row, col
```

```python
def utility_func_cd(ga,gb,m,parameter):
  a=ga**(parameter[0]-1)

  b=gb**(parameter[1]-1)

  c=m**(parameter[2]-1)

  utility=a*b*c
  utility=float(utility)
  return utility
```

```python
def utility_func_s(ga,gb,m,parameter):
  b=(ga+gb)**(parameter[0]-1)

  

  a=m**(parameter[2]-1)

  utility=a*b
  utility=float(utility)
  return utility
```

```python
def utility_func_c(ga,gb,m,parameter):
  a=m**(parameter[2]-1)

  utility=a*(min(ga,gb)**(parameter[0]-1))
  utility=float(utility)
  return utility
```

## Metroplolis

```python
# define a function to sample
def random_pick(g, G, a):
    # Randomly select a number within the range [g-mg, g+mg]
    delta_g=G*a
    new_g = random.uniform(g - delta_g, g + delta_g)

    # Apply the reflex logic
    if new_g < 0:
        new_g = - new_g
    elif new_g > G:
        new_g = 2*G - new_g

    return new_g
```

```python
def random_pickint(g, G1, a):
    # Randomly select a number within the range [g-mg, g+mg]
    delta_g=round(G1*a)
    new_g = random.randint(g - delta_g, g + delta_g)

    # Apply the reflex logic, if g picked is out of range, then take the reflex value, this makes
    # See the metroplis formula

    if new_g < 0.5:
        new_g = 1- new_g
    elif new_g > G1-0.5:
        new_g = 2*(G1-0.5) - new_g
    if new_g==0 or new_g==G1:
      print('Error',G1,g,delta_g)
    return new_g
```
